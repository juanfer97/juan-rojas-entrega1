# 01 - exploración
Objetivo: cargar `train.csv` de la competencia y mostrar una exploración básica de los datos.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import files
uploaded = files.upload()   # selecciona train.csv cuando aparezca el diálogo
fname = next(iter(uploaded))
df = pd.read_csv(fname)
print("Archivo cargado:", fname)
```

```python
print("Forma del dataset:", df.shape)
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
if 'target' in df.columns:
    print(df['target'].value_counts())
elif 'clase' in df.columns:
    print(df['clase'].value_counts())
else:
    print("No encontré columna 'target' ni 'clase'. Columnas disponibles:")
    print(df.columns.tolist())
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing[missing > 0]
```

```python
for col in ['localidad','region','ciudad']:
    if col in df.columns:
        display(df[col].value_counts().head(20))
```

```python
num_cols = df.select_dtypes(include=['number']).columns.tolist()
if len(num_cols) > 0:
    df[num_cols[0]].hist(bins=20)
    plt.title(f'Histograma de {num_cols[0]}')
    plt.xlabel(num_cols[0])
    plt.show()

if 'target' in df.columns:
    df['target'].value_counts().plot(kind='bar')
    plt.title('Distribución de target')
    plt.show()
```

```python
if 'clase' in df.columns:
    df['clase'].value_counts().plot(kind='bar', figsize=(6,4))
    plt.title('Distribución de clases')
    plt.xlabel('Clase')
    plt.ylabel('Cantidad de estudiantes')
    plt.show()
```

```python
# Conteo de estudiantes por clase (rendimiento global)
conteo_clases = df["RENDIMIENTO_GLOBAL"].value_counts()
print("Número de estudiantes por clase:\n")
print(conteo_clases)

# Conteo de estudiantes por departamento
conteo_departamento = df["E_PRGM_DEPARTAMENTO"].value_counts()
print("\nNúmero de estudiantes por departamento:\n")
print(conteo_departamento)

# Gráfico de barras del rendimiento global
plt.figure(figsize=(6,4))
conteo_clases.plot(kind='bar', color='skyblue')
plt.title("Distribución de estudiantes por clase (rendimiento global)")
plt.xlabel("Clase")
plt.ylabel("Número de estudiantes")
plt.show()

# Gráfico de barras de los departamentos con más estudiantes
plt.figure(figsize=(8,4))
conteo_departamento.head(10).plot(kind='bar', color='lightgreen')
plt.title("Top 10 departamentos con más estudiantes")
plt.xlabel("Departamento")
plt.ylabel("Número de estudiantes")
plt.show()
```

### Observaciones

Al analizar la variable de rendimiento global se observa que la mayoría de los estudiantes se encuentran en la categoría **“alto”**, con más de 175 mil registros. Le siguen muy de cerca las categorías **“bajo”**, **“medio-bajo”** y **“medio-alto”**, todas con cifras similares. Esto sugiere que el rendimiento de los estudiantes está bastante equilibrado, sin una diferencia marcada entre los grupos.

En cuanto a los departamentos, se nota una gran concentración de estudiantes en **Bogotá**, que reúne más de 280 mil casos. Luego aparecen **Antioquia**, **Valle** y **Atlántico** con cantidades considerablemente menores, pero que siguen siendo representativas. Los demás departamentos tienen una participación mucho más baja, e incluso algunos como **Vaupés**, **San Andrés** o **Amazonas** apenas registran unos pocos estudiantes.

En general, esta exploración inicial permite entender mejor la composición de la base de datos y muestra que la mayor parte de los registros proviene de regiones con alta concentración urbana y educativa, lo cual podría influir en los resultados de rendimiento.
